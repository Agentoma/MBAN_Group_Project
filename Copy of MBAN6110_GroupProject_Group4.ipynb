# REDUCING FEATURES

```python
# Combining Heart Attack and Heart Stroke
df['HadHeartAttackOrStroke'] = df['HadHeartAttack'] + df['HadStroke']
df['HadHeartAttackOrStroke'] = df['HadHeartAttackOrStroke'].where(df['HadHeartAttackOrStroke'] <= 1, other=1)
df['HadHeartAttackOrStroke'].value_counts(dropna=False)
```

```python
df.drop(columns=['HadHeartAttack', 'HadStroke'], inplace=True)
df.info()
```

```python
# Combining Physical Disabilities as Quantity
df['PhysicalDisabilities'] = df['DeafOrHardOfHearing'] + df['BlindOrVisionDifficulty'] + df['DifficultyConcentrating'] + df['DifficultyWalking'] + df['DifficultyDressingBathing'] + df['DifficultyErrands']
df['PhysicalDisabilities'].value_counts(dropna=False)
```

```python
df = df.drop(columns=['DeafOrHardOfHearing', 'BlindOrVisionDifficulty', 'DifficultyConcentrating', 'DifficultyWalking', 'DifficultyDressingBathing', 'DifficultyErrands'])
df.info()
```

```python
# Combine smoker status and ecig usage (weighted by smoker status having greater impact)
df['SmokerRating'] = df['SmokerStatus'] + df['ECigaretteUsage']/2
df['SmokerRating'].value_counts(dropna=False)
```

```python
df.drop(columns=['SmokerStatus', 'ECigaretteUsage'], inplace=True)
```

```python
# Remove redundant columns (weight and height vs BMI)
df = df.drop(columns=['WeightInKilograms','HeightInMeters'])
```

```python
df.info()
```

```python
# Combine lung diseases
df['HadLungDisease'] = df['HadAsthma'] + df['HadCOPD']
df['HadLungDisease'] = df['HadLungDisease'].where(df['HadLungDisease'] <= 1, other=1)
df.drop(columns=['HadAsthma', 'HadCOPD'], inplace=True)
```
